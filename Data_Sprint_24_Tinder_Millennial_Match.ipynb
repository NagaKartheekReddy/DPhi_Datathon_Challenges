```
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import OneHotEncoder
from sklearn.ensemble import RandomForestClassifier
```

```
train_data = pd.read_csv("https://raw.githubusercontent.com/dphi-official/Datasets/master/Tinder_Millennial_Match/train_set_label.csv")
train_data
```

```
train_data.dtypes
```

```
category2vec = train_data.select_dtypes('object')
category2vec
```

```
ohe = OneHotEncoder(handle_unknown = 'ignore')
encoded_train = ohe.fit_transform(category2vec).toarray()
train = pd.DataFrame(encoded_train, columns=ohe.get_feature_names(category2vec.columns))
train
```

```
train_data = train_data.drop(category2vec.columns, axis=1)
train_data = pd.concat([train_data, train], axis=1)
train_data
```

```
X = train_data.drop(['It became a relationship'], axis=1)
Y = train_data['It became a relationship']
```

```
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=42)
```

```
poly = RandomForestClassifier(n_estimators=100, oob_score = True, n_jobs = -1, random_state=50).fit(X_train, Y_train)
poly_pred = poly.predict(X_test)

poly_accuracy = accuracy_score(Y_test, poly_pred)

print(poly_accuracy)
```

```
test_data = pd.read_csv('https://raw.githubusercontent.com/dphi-official/Datasets/master/Tinder_Millennial_Match/test_set_label.csv')
test_data
```

```
category2vec_test = test_data.select_dtypes('object')
category2vec_test
```

```
encoded_test = ohe.transform(category2vec_test).toarray()
test = pd.DataFrame(encoded_test, columns=ohe.get_feature_names(category2vec_test.columns))
```

```
test_data = test_data.drop(category2vec_test.columns, axis = 1)
test_data = pd.concat([test_data, test], axis = 1)
```

```
target = poly.predict(test_data)
res = pd.DataFrame(target)
res.columns = ["prediction"]
from google.colab import files
res.to_csv('submission.csv', index=False)
files.download('submission.csv')
```
